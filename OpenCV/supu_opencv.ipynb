```python
#opencv-pythonのインポート
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img_1.png')
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
cv2.imwrite('./output.png', img)
```

```python
resize_img = cv2.resize(img, (500, 200))
plt.imshow(cv2.cvtColor(resize_img, cv2.COLOR_BGR2RGB))
```

```python
h = round(img.shape[0]/2)
w = round(img.shape[0]/2)
small_img = cv2.resize(img, (w, h))
plt.imshow(cv2.cvtColor(small_img, cv2.COLOR_BGR2RGB))
```

```python
trim_img = small_img[40:100, 40:110]
plt.imshow(cv2.cvtColor(trim_img, cv2.COLOR_BGR2RGB))
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB))
```

```python
rotate_img = cv2.rotate(img,cv2.ROTATE_180)
plt.imshow(cv2.cvtColor(rotate_img, cv2.COLOR_BGR2RGB))
```

```python
img_flip = cv2.flip(img, 1)
plt.imshow(cv2.cvtColor(img_flip, cv2.COLOR_BGR2RGB))
```

```python
img_abs = cv2.convertScaleAbs(img, alpha=1.0, beta = 80)
plt.imshow(cv2.cvtColor(img_abs, cv2.COLOR_BGR2RGB))
```
